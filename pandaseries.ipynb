For this dataset, get following answers,
1. What was the maximum temperature in each of these 3 cities?

```python
def grouper(data, idx, col):
    if 80 <= data[col].loc[idx] <= 90:
        return '80-90'
    elif 50 <= data[col].loc[idx] <= 60:
        return '50-60'
    else:
        return 'others'
```

```python
g=new.groupby(lambda x:grouper(new,x,'temperature'))
g
```

```python
for Range, info in g: #key is return value of fuunction
    print("Temp group:{}\n".format(Range))
    print(info)
    print("\n")
```

## merge

```python
first=pd.DataFrame({'stocks':['shoes','goggle','pant','shirt'],"price(RS)":[800,250,590,350]})
first
```

```python
second=pd.DataFrame({'stocks':['shoes','goggle','pant','shirt'],'brand':['amazon','flipkart','myntra','snapdeal']})
second
```

```python
add=pd.merge(first,second,on="stocks")
add
```

```python
third=pd.DataFrame({'stocks':['shoes','pocket','pant','books'],'brand':['amazon','flipkart','myntra','snapdeal']})
third
```

```python
add2=pd.merge(first,third,on='stocks',how="inner") #on the basis of stocks and comman  
add2                                               #default valuee of how=inner   
```

```python
add3=pd.merge(first,third,on='stocks',how="outer") # missing value with nan
add3
```

```python
dd2=pd.merge(first,third,on='stocks',how="left") #left means stocks of only first
dd2
```

```python
dd2=pd.merge(first,third,on='stocks',how="right") #left means stocks of only second
dd2
```

```python
dd2=pd.merge(first,third,on='stocks',how="outer",indicator=True) #indicate stocks from which frame
dd2
```
