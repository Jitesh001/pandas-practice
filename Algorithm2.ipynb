```python
def longest_substring(x,y):
    n,m = (len(y)+1),(len(x)+1) #m= columns, n =rows
    lis = [[0 for i in range(m)] for j in range(n)]
    for i in range(n):
        for j in range(m):
            if i == 0 or j == 0:
                lis[i][j] = 0
            elif y[i-1] == x[j-1]: # if same increse diagonal value
                lis[i][j] = lis[i-1][j-1] + 1
            else:
                lis[i][j] = max(lis[i][j-1], lis[i-1][j])
            
    return lis[n-1][m-1] ##lenght of substring
print(longest_substring('AGGTAB','GXTXAYB'))
print(longest_substring('AABBBABABABAB', "CSGSAKSJBSJSAAJSB"))
```

```python
## edit distance using just insert , modify and delete operation
## calculate distance(no. of operations) to convert str1 to str2

def totalOp(str1, str2):
    n,m = len(str1)+1, len(str2)+1  #n -rows and m-columns
    lis = [[0 for j in range(m)] for i in range(n)]
    for i in range(n):
        for j in range(m):
            if i == 0:
                lis[i][j] = j
            elif j == 0:
                lis[i][j] = i
            elif str1[i-1] == str2[j-1]:
                lis[i][j] = lis[i-1][j-1]
            else:
                lis[i][j] = 1 + min(lis[i][j-1], lis[i-1][j], lis[i-1][j-1])
    return lis[n-1][m-1]
print(totalOp('CART', 'MARCH'))
```

```python
## bionomial coefficent C(n, K) = C(n-1, k)+C(n-1, k-1)
## recursive methods lets use DP

## using 2D array
def bino_2D(n,k):
    lis = [[0 for i in range(k+1)] for j in range(n+1)]
    for i in range(n+1):
        for j in range(min(i,k)+1):
            if j == 0 or j == i:
                lis[i][j] = 1
            else:
                lis[i][j] = lis[i-1][j-1] + lis[i-1][j]
    return lis[n][k]
print("using 2D array :")
print(bino_2D(4,2))


## using 1D array
def bino_1D(n,k):
    lis = [1 if i == 0 else 0 for i in range(k+1)]
    for i in range(n+1):
        for j in range(min(i, k), 0,-1):
            lis[j] = lis[j] + lis[j-1]
    return lis[-1]
print("using 1D array :")
print(bino_1D(4,2))
```

```python
#knapSack 0-1

def knapSack(values, weights, W):
    n =len(values)
    lis = [[0 for i in range(W+1)] for j in range(n+1)]
    for i in range(n+1):
        for j in range(W+1):
            if i==0 or j== 0:
                lis[i][j] = 0
            elif weights[i-1] <=j:
                lis[i][j] = max(values[i-1], lis[i-1][j-weights[i-1]], lis[i-1][j])
            else:
                lis[i][j] = lis[i-1][j]
    return lis[n][W]
values = [60,100,120]
weights = [10,20,30]
W = 50
print(knapSack(values, weights, W))
```

### Searching algorithm

Linear Search:
- Traverse the list or array sequentially and compare each element with the target until a match is found or the end of the list is reached.
- Time complexity: O(n) in the worst case.

```python
lis = [4,6,3,9,20,13]
find = 10
for elem in lis:
    if elem == find:
        print('found')
        break
else:
    print('not found')
```

Binary Search:
- Applicable to sorted arrays.
- Divide the array in half and compare the target with the middle element. If the target is smaller, repeat the search on the left half; if larger, on the right half.
- Time complexity: O(log n) in the worst case.

```python
arr = [4,5,6,8,9,20,48,49,84]
target = 48 
found = False
left, right = 0, len(arr)-1

while left <= right:
    mid = (left+right)//2 
    
    if arr[mid] == target:
        found = True
        index = mid
        break 
    
    if arr[mid] < target:
        left = mid + 1 
    else:
        right = mid - 1 
    
if found:
    print(f'element found at {index}')
else:
    print('No element')
```

Interpolation Search:
- An enhancement of binary search for uniformly distributed data.
- Uses an estimate to guess the likely position of the target based on the value range.
- Time complexity: O(log log n) on average for uniformly distributed data.

```python
# Example Usage:
arr = [2, 5, 8, 12, 16, 23, 38, 42]
target = 100

# Ensure the array is sorted before using interpolation search
arr.sort()

low, high = 0, len(arr)-1

while low <= high and (arr[low] <= target <= arr[high]):
    # Calculate the position of the target element
    pos = low + ((target - arr[low]) * (high - low)) // (arr[high] - arr[low])

    if arr[pos] == target:
        print('found at index', pos)  # Element found, return the index
        break
    elif arr[pos] < target:
        low = pos + 1  # Search the right half
    else:
        high = pos - 1  # Search the left half

else:
    print('not found')
```

### Sorting algos

Bubble Sort:
- Repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order.
-Time complexity: O(n^2) in the worst case.

```python
arr = [64, 34, 25, 12, 22, 11, 90] 
n = len(arr)

for i in range(n):
    for j in range(n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
            
print(arr)
```

Selection Sort:
- Divides the list into a sorted and an unsorted region. In each iteration, it selects the smallest (or largest) element from the unsorted region and swaps it with the first (or last) element of the unsorted region.
- Time complexity: O(n^2) in the worst case.

```python
arr = [64, 34, 25, 12, 22, 11, 90]
n = len(arr)

# Traverse through all array elements
for i in range(n):
    # Find the minimum element in the remaining unsorted array
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    # Swap the found minimum element with the first element
    arr[i], arr[min_index] = arr[min_index], arr[i]
    
print(arr)
```

Insertion Sort:
- Builds the final sorted array one element at a time. It takes each element from the list and inserts it into its correct position in the sorted part of the array.
- Time complexity: O(n^2) in the worst case.

```python
#insertion sort
arr = [64, 34, 25, 12, 22, 11, 90]
n = len(arr)
for i in range(1, n):
    j = i-1
    elem = arr[i]
    while j >= 0 and elem < arr[j]:
        arr[j+1] = arr[j]
        j -= 1
    
    arr[j+1] = elem
        
            
print(arr)
```

Merge Sort:
- Divides the unsorted list into n sub-lists, each containing one element, and then repeatedly merges sub-lists to produce new sorted sub-lists until there is only one sub-list remaining.
- Time complexity: O(n log n) in the worst case.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Find the middle of the list
        left_half = arr[:mid]  # Divide the list into two halves
        right_half = arr[mid:]

        # Recursive call on each half
        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        # Merge the two halves back together
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Check for any remaining elements in the left and right halves
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr

# Example Usage:
unsorted_list = [64, 34, 25, 12, 22, 11, 90]
print("Original List:", unsorted_list)
sorted_list = merge_sort(unsorted_list)
print("Sorted List:", sorted_list)
```

```python
def gcd_iterative(a, b):
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a

print(gcd_iterative(4,6))

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(gcd(4,6))
```

```python
def lcm(a, b):
    max_num = max(a, b)
    min_num = min(a, b)

    current_multiple = max_num

    while True:
        if current_multiple % min_num == 0:
            return current_multiple
        current_multiple += max_num

print(lcm(4,5))
```

```python
def is_valid_expression(expression):
    stack = []
    opening_brackets = "({["
    closing_brackets = ")}]"

    for char in expression:
        if char in opening_brackets:
            stack.append(char)
        elif char in closing_brackets:
            if not stack or opening_brackets[closing_brackets.index(char)] != stack.pop():
                return False
        
        
    return not stack  # True if the stack is empty

# Example usage:
expression1 = "{(a + b) * [c - d]}"
expression2 = "{(a + b) * [c - d}}"
expression3 = "((a + b) * c - d)"

print(is_valid_expression(expression3))  # Output: True
```

```python
def matched(a, b):
    opening = '([{'
    closing = ')]}'
    
    return opening.index(a) == closing.index(b)
    
def fun(str1):
    lis = []
    i = 0
    balance = True
    while i < len(str1) and balance:
        symb = str1[i]
        if symb not in '([{)]}':
            i+=1
            continue
        if symb in '([{':
            lis.append(symb)
        else:
            if not lis:
                balance = False
            else:
                top = lis.pop()
                if not matched(top, symb):
                    balance = False
        i+=1 
   

    if not lis and balance:
        return True
    
    return False

expression3 = "((a + b) * c - d)"
print(fun(expression3))
```
