```python
from datetime import datetime

def is_date_valid(date_string):
    try:
        # Try to parse the date string
        datetime.strptime(date_string, "%Y-%m-%d")
        return True
    except ValueError:
        # The date is not valid
        return False
```

```python
import pandas as pd

# Create two DataFrames
data1 = {'ID': [1, 2, 3],
         'Name': ['John', 'Jane', 'Sam']}
df1 = pd.DataFrame(data1)

data2 = {'ID': [2, 3, 4],
         'Age': [25, 30, 35]}
df2 = pd.DataFrame(data2)

# Inner join based on 'ID' column
df_inner = pd.merge(df1, df2, on='ID', how='inner')
df_inner
```

```python
df_left = pd.merge(df1, df2, on='ID', how='left')
df_left
```

```python
df_right = pd.merge(df1, df2, on='ID', how='right')
df_right
```

```python
df_outer = pd.merge(df1, df2, on='ID', how='outer')
df_outer
```

stack() function:
The stack() function is used to pivot a level of column labels into the innermost level of the index, resulting in a "long" format DataFrame. It compresses the column labels into a single level of the index, making the DataFrame taller.

unstack() function:
The unstack() function is used to pivot a level of the index labels into the column labels, resulting in a "wide" format DataFrame. It expands a level of the index to become the column labels, making the DataFrame wider

```python
import pandas as pd

# Create a sample DataFrame with hierarchical index
data = {
    ('Group 1', 'A'): [1, 2, 3],
    ('Group 1', 'B'): [4, 5, 6],
    ('Group 2', 'A'): [7, 8, 9],
    ('Group 2', 'B'): [10, 11, 12]
}

df = pd.DataFrame(data)
print(df)
# Stack the DataFrame
stacked = df.stack()
print(stacked)
```

```python
# Unstack the DataFrame
unstacked = stacked.unstack()
print(unstacked)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Value1': [10, 20, 30, 40, 50, 60],
        'Value2': [100, 200, 300, 400, 500, 600]}
df = pd.DataFrame(data)

# Group by the 'Category' column
grouped = df.groupby('Category')

# Apply different aggregation functions to different columns
result = grouped.agg({
    'Value1': 'sum',       # Sum of 'Value1'
    'Value2': ['mean', 'max']  # Mean and max of 'Value2'
})

print(result)
```
